**Tensors (Collection of homogeneous type of data)**


---


1. Tensor is a collection of same type of multidimentional matrix

2. Similar to NumPy array but more powerfull to work on GPUs rather than CPUs
3. Its default datatype is float32
4. it is more suitable for deep learning, GenAI and Agentic AI    than Numpy array
5. allows negative as well as positive indexing

```python
import torch as tr
import numpy as np
```
